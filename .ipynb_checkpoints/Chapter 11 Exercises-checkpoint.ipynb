## 1. Programming Exercises - Exercise 1 

**1. Employee and ProductionWorker Classes**

Write an *Employee* class that keeps data attributes for the following pieces of information:

* Employee name
* Employee number

Next, write a class named *ProductionWorker* that is subclass of the *Employee* class. The *ProductionWorker* class should keep data attributes for the following information: 

* Shift number (an integer, such as 1, 2, or 3) 
* Hourly pay rate 

**The workday is divided into shift days (day and night):** The shift attribute will hold an integer value representing the shift that the employee works. The day shift is 'shift 1' and the night shift is 'shift 2'. Write the appropriate accessor and mutator methods for each class.

Once you have written the classes, write a program that creates an object of the *ProductionWorker* class and prompts the user to enter data for each of the object's data attributes. Store the data in the object, then use the object's accessor methods to retrieve it and display it on the screen. 

**Program Description:** The program begins by defining the parent class named 'Employee' which includes a constructor method that initializes an instance of the 'Employee' class with both the attributes 'name' and number'. The associated set methods prompt the user to input data for each attribute, and afterwards the get methods are then utilized for retrieving and displaying these values.

Next, the subclass (child class) 'Production worker' is defined, it inherits the attributes 'name' and 'number' by referencing the parent class 'Employee' and calling the constructor method from the Employee class to handle its attributes. The subclass also uses an 'init' constructor method to create an instance of itself with the inherited attributes and its own 'shift' and 'pay' attributes. 

Set methods (set_shift and set_pay) will retrieve user input for the attributes, specifically 'self.shift' utilizes an infinite while loop that'll continuously prompt the user to enter the number '1' or '2' through the 'else' statement (referencing shift 1 and shift 2) until one of these numbers are chosen, in which the respective if or elif statement will reassign the new value with a string (stating the type of shift and number) before breaking the loop and continuing the program. 

Afterwards the get methods will retrieve and display these values (once called). The program will then create an object named 'worker' that serves as an instance of the 'ProductionWorker' class, the attributes of the class are initialized with empty strings that'll serve as placeholders before the set methods are called and the values of each attribute are reassigned based on user input, the values are then displayed through a print statement using the get methods. 

```python
# Function is defined as 'main' and serves as the starting point of the program
def main():

# The parent class 'Employee' is initialized and is associated with the 'name' and 'number'
# attributes
    class Employee:

# '__init__' constructor method is used to initialize an instance of 'Employee' with the 'name'
# 'number' attributes.
        def __init__(self, name, number):
            self.name = name
            self.number = number

# The method'set_name' prompts the user to input the Employee's name, this
# value is then assigned to the 'name' attribute of the 'Employee' instance
        def set_name(self, name):
            self.name = input("Enter your name here: ")

# The method 'set_number' prompts the user to input the Employee's number, this
# value is then assigned to the 'number' attribute of the 'Employee' instance    
        def set_number(self, number):
            self.number = input("Enter your Employee Number here: ")

# The method 'get_name' is defined and uses 'return self.name' to retrieve and return
# the value for the 'name' attribute of the 'Employee' instance
        def get_name(self):
            return self.name

# The method 'get_number' is defined and uses 'return self.name' to retrieve and return
# the value for the 'number' attribute of the 'Employee' instance
        def get_number(self):
            return self.number

# The child class 'ProductionWorker' is created, this being a subclass of 'Employee'
# which inherits its 'name' and 'number' attributes
    class ProductionWorker(Employee):

# '__init__' constructor method is used to initialize an instance of 'ProductionWorker' with the 'name'
# 'number' attributes that it inherited from 'Employee', as well as its own 'shift' and pay attributes.
        def __init__(self, name, number, shift, pay):

# Employee.__init__ calls the constructor of the Employee class to handle the 'name' and 'number'
# attributes. self.shift = shift and self.pay = pay assign the values of the parameters
# to the instance variables 
            Employee.__init__(self, name, number)
            self.shift = shift
            self.pay = pay

# The 'set_shift' method is defined within 'ProductionWorker', the 'shift' parameter is used to
# set the shift attribute of the instance
        def set_shift(self, shift):

# The 'while True' statement creates an infinite loop as the 'True' condition never becomes 'False' and
# by default it's True, thus repeating its contents until the 'break' statement is executed
            while True:
        
# The user is prompted to enter either "1" or "2" to represent their shift number. The user's input is then
# stored in 'self.shift'
                self.shift = input("Enter your shift number (1 or 2): ")

# The if statement checks if the value of 'self.shift' is '1', If this is true 'self-shift' is then reassigned 
# a string (replacing its value) that states the Employee's shift, that being the 'Day Shift' and 'Shift 1'. 
# The 'break' statement then exits the while loop.
                if self.shift in ["1"]:
                    self.shift = "Employee's Shift: 'Day Shift - Shift 1'"        
                    break

# The elif statement checks if the value of 'self.shift' is '2', If this is true 'self-shift' is then reassigned 
# a string (replacing its value) that states the Employee's shift, that being the 'Night Shift' and 'Shift 2'. 
# The 'break' statement then exits the while loop.            
                elif self.shift in ["2"]:
                    self.shift = "Employee's Shift: 'Night Shift - Shift 2'"
                    break

# The 'else' statement is executed if the value of 'self.shift' is neither '1' or '2', if this occurs a print
# statement displays a string instructing the user to enter a valid number, after which the while loop will repeat, 
# prompting the user to enter their shift number once again.
                else:
                    print("[Please enter a valid number]")

# The method 'set_pay' prompts the user to input the Production Worker's hourly salary, this
# value is then assigned to the 'pay' attribute of the 'ProductionWorker' instance       
        def set_pay(self, pay):
            self.pay = input("Enter your hourly salary: ")

# The method 'get_shift' is defined and uses 'return self.shift' to retrieve and return
# the value for the 'shift' attribute of the 'ProductionWorker' instance
        def get_shift(self):
            return self.shift

# The method 'get_pay' is defined and uses 'return self.pay' to retrieve and return
# the value for the 'pay' attribute of the 'ProductionWorker' instance    
        def get_pay(self):
            return self.pay

# The variable worker (the object) creates an instance of the 'ProductionWorker' class, including its
# attributes and those inherited from the 'Employee' parent class, these attributes are
# initialized with empty strings as placeholders
    worker = ProductionWorker("","","","")

# Print statement displays a string to informing the user to submit Employee information, preparing them
# for input prompts 
    print("[Please submit Employee information below]")

# Each set method is called from the 'worker' object with an empty string that's passed as an 
# argument, serving as a placeholder value 
    worker.set_name("")
    worker.set_number("")
    worker.set_shift("")
    worker.set_pay("")

# The empty print statement serves as a line break to give the output a cleaner presentation 
    print()

# Print statement displays a string that informs the user of the 'Employee Information' before
# the information is presented 
    print("[Employee Information]")

# The multiple print statements prefixed with an f-string display the information along with calling the 'get methods'
# which also display the values of each variable belonging to the 'worker' object
    print(f"Name of Employee: {worker.get_name()}")
    print(f"Employee's Work Number: {worker.get_number()}")
    print(worker.get_shift())
    print(f"Employee's Hourly Pay: {worker.get_pay()}")
    
# The main function will only be called if this script is ran directly    
if __name__ == "__main__":
    main()
```

## 2. Programming Exercises - Exercise 2

**2. ShiftSupervisor Class**

In a particular factory, a shift supervisor is a salaried employee who supervises a shift. In addition to a salary, the shift supervisor earns a yearly bonus when his or her shift meets production goals. Write a *ShiftSupervisor* class that is a subclass of the *Employee* class you created in Programming Exercise 1. 

The *ShiftSupervisor* class should keep a data attribute for the annual salary, and a data attribute for the annual production bonus that a shift supervisor has earned. Demonstrate the class by writing a program that uses a *ShiftSupervisor* object.

**Program Description:** Since this Programming Exercise is basically an extension to the first, i'll explain the new things i've added. For this, the subclass 'ShiftSupervisor' is defined, it inherits attributes from the 'Employee' class and has its own additional attributes, that being 'shift' and 'pay'. Within the class the 'init' method initializes an instance of the 'ShiftSupervisor' class, taking the attributes as parameters. 

The set methods are then used to assign values to the 'annual' and 'bonus attributes through user input (when called) and the get methods will retrieve and return the data as visual output to the user (when called). The object 'supervisor' creates an instance of 'ShiftSupervisor' with empty strings as placeholders before the values of the attributes are reassigned using the set methods. 

Once the user has input the information, the program displays the values of these attributes back to the user. 

```python
# Function is defined as 'main' and serves as the starting point of the program
def main():

# # The parent class 'Employee' is initialized and is associated with the 'name' and 'number'
# attributes  
    class Employee:

# '__init__' constructor method is used to initialize an instance of 'Employee' with the 'name'
# 'number' attributes.    
        def __init__(self, name, number):
            self.name = name
            self.number = number

# The method 'set_name' prompts the user to input the Employee's name, this
# value is then assigned to the 'name' attribute of the 'Employee' instance  
        def set_name(self, name):
            self.name = input("Enter your name here: ")

# The method 'set_number' prompts the user to input the Employee's number, this
# value is then assigned to the 'number' attribute of the 'Employee' instance     
        def set_number(self, number):
            self.number = input("Enter your Employee Number here: ")

# The method 'get_name' is defined and uses 'return self.name' to retrieve and return
# the value for the 'name' attribute of the 'Employee' instance  
        def get_name(self):
            return self.name

# The method 'get_number' is defined and uses 'return self.name' to retrieve and return
# the value for the 'number' attribute of the 'Employee' instance
        def get_number(self):
            return self.number

# The child class 'ProductionWorker' is created, this being a subclass of 'Employee'
# which inherits its 'name' and 'number' attributes  
    class ProductionWorker(Employee):

# '__init__' constructor method is used to initialize an instance of 'ProductionWorker' with the 'name'
# 'number' attributes that it inherited from 'Employee', as well as its own 'shift' and pay attributes.    
        def __init__(self, name, number, shift, pay):
        
# Employee.__init__ calls the constructor of the Employee class to handle the 'name' and 'number'
# attributes. self.shift = shift and self.pay = pay assign the values of the parameters
# to the instance variables 
            Employee.__init__(self, name, number)
            self.shift = shift
            self.pay = pay

# The 'set_shift' method is defined within 'ProductionWorker', the 'shift' parameter is used to
# set the shift attribute of the instance  
        def set_shift(self, shift):

# The 'while True' statement creates an infinite loop as the 'True' condition never becomes 'False' and
# by default it's True, thus repeating its contents until the 'break' statement is executed
            while True:
        
# The user is prompted to enter either "1" or "2" to represent their shift number. The user's input is then
# stored in 'self.shift'
                self.shift = input("Enter your shift number (1 or 2): ")

# The if statement checks if the value of 'self.shift' is '1', If this is true 'self-shift' is then reassigned 
# a string (replacing its value) that states the Employee's shift, that being the 'Day Shift' and 'Shift 1'. 
# The 'break' statement then exits the while loop.
                if self.shift in ["1"]:
                    self.shift = "Employee's Shift: 'Day Shift - Shift 1'"        
                    break

# The elif statement checks if the value of 'self.shift' is '2', If this is true 'self-shift' is then reassigned 
# a string (replacing its value) that states the Employee's shift, that being the 'Night Shift' and 'Shift 2'. 
# The 'break' statement then exits the while loop.   
                elif self.shift in ["2"]:
                    self.shift = "Employee's Shift: 'Night Shift - Shift 2'"
                    break

# The 'else' statement is executed if the value of 'self.shift' is neither '1' or '2', if this occurs a print
# statement displays a string instructing the user to enter a valid number, after which the while loop will repeat, 
# prompting the user to enter their shift number once again.
                else:
                    print("[Please enter a valid number]")

# The method 'set_pay' prompts the user to input the Production Worker's hourly salary, this
# value is then assigned to the 'pay' attribute of the 'ProductionWorker' instance       
        def set_pay(self, pay):
            self.pay = input("Enter your hourly salary: ")

# The method 'get_shift' is defined and uses 'return self.shift' to retrieve and return
# the value for the 'shift' attribute of the 'ProductionWorker' instance  
        def get_shift(self):
            return self.shift

# The method 'get_pay' is defined and uses 'return self.pay' to retrieve and return
# the value for the 'pay' attribute of the 'ProductionWorker' instance    
        def get_pay(self):
            return self.pay

# The second child class 'ShiftSupervisor' is created, this is another subclass of 'Employee'
# which inherits its 'name' and 'number' attributes just like the 'ProductionWorker' class   
    class ShiftSupervisor(Employee):

# '__init__' constructor method is used to initialize an instance of 'ShiftSupervisor' with the 'name'
# 'number' attributes that it inherited from 'Employee', as well as its own 'annual' and 'bonus' attributes.
        def __init__(self, name, number, annual, bonus):
            Employee.__init__(self, name, number)
            self.annual = annual
            self.bonus = bonus

# The method 'set_annual' prompts the user to input the Shift Supervisor's annual salary, this
# value is then assigned to the 'annual' attribute of the 'ShiftSupervisor' instance    
        def set_annual(self, annual):
            self.annual = input("Enter your annual salary: ")

# The method 'set_bonus' prompts the user to input the Shift Supervisor's annual production bonus, 
# this value is then assigned to the 'bonus' attribute of the 'ShiftSupervisor' instance
        def set_bonus(self, bonus):
            self.bonus = input("Enter your annual production bonus: ")

# The method 'get_annual' is defined and uses 'return self.annual' to retrieve and return
# the value for the 'annual' attribute of the 'ShiftSupervisor' instance   
        def get_annual(self):
            return self.annual

# The method 'get_bonus' is defined and uses 'return self.bonus' to retrieve and return
# the value for the 'bonus' attribute of the 'ShiftSupervisor' instance   
        def get_bonus(self):
            return self.bonus

# The variable supervisor (the object) creates an instance of the 'ShiftSupervisor' class, including its
# attributes and those inherited from the 'Employee' parent class, these attributes are
# initialized with empty strings as placeholders
    supervisor = ShiftSupervisor("","","","")

# Print statement displays a string to informing the user to submit Employee information, preparing them
# for input prompts  
    print("[Please submit Employee information below]")

# Each set method is called from the 'supervisor' object with an empty string that's passed as an 
# argument, serving as a placeholder value 
    supervisor.set_name("")
    supervisor.set_number("")
    supervisor.set_annual("")
    supervisor.set_bonus("")

# The empty print statement serves as a line break to give the output a cleaner presentation   
    print()

# Print statement displays a string that informs the user of the 'Employee Information' before
# the information is presented 
    print("[Employee Information]")

# The multiple print statements prefixed with an f-string display the information along with calling the 'get methods'
# which also display the values of each variable belonging to the 'supervisor' object
    print(f"Name of Employee: {supervisor.get_name()}")
    print(f"Employee's Work Number: {supervisor.get_number()}")
    print(f"Employee's Annual Salary: {supervisor.get_annual()}")
    print(f"Employee's Annual Production Bonus: {supervisor.get_bonus()}")

# The main function will only be called if this script is ran directly    
if __name__ == "__main__":
    main()
```
