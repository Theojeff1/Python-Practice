**1. Pet Class**

**Write a class named Pet, which should have the following data attributes:**

* __name (for the name of a pet) <br>
* __animal_type (for the type of animal that a pet is. Example values are ‘Dog’, ‘Cat’,and ‘Bird’) <br>
* __age (for the pet’s age) <br>

**The Pet class should have an __init__ method that creates these attributes. It should also have the following methods:**

* set_name <br>
(This method assigns a value to the __name field.) 

* set_animal_type <br>
(This method assigns a value to the __animal_type field)

* set_age <br>
(This method assigns a value to the __age field)

* get_name <br>
(This method returns the value of the __ name field)

* get_ animal_type <br>
(This method returns the value of the _animal_type field)

* get_age <br>
(This method returns the value of the _age field)

Once you have written the class, write a program that creates an object of the class and prompts the user to enter the name, type, and age of his or her pet. This data should be stored as the object’s attributes. Use the object’s accessor methods to retrieve the pet’s name, type, and age and display the data on the screen.

**Program Description:** This program begins by defining the class for this exercise which this class is labeled as 'Pet', it is subsequently followed by the constructor method 'init' this initializes an instance of the Pet class *(used to help set up values for the variables)* that has the parameters 'name', 'animal_type', 'age'. The values of these parameters are then assigned and linked to the matching attributes of 'Pet' *(so this way the values are stored and accessible)*. 

The functions set_name, set_animal_type, and set_age are then created for the purpose of acquiring user input *(when called)* to get the values which are the assigned to their respective attributes. More functions are defined which are get_name, get_animal_type, and get_age which *(when called)* return their provided string and current values as output. 

The object 'my_pet' is created, its empty strings are passed as arguments to the constructor's parameter which initialize the attributes with empty values *(these will serve as placeholders)*. Now that everything is set up, the functions that were defined earlier can be called, they're provided the input values of the attributes *(ex: set_name)* and are then displayed as output for the user to see *(ex: get_name)*.

```python
# Function is defined as 'main' and serves as the starting point of the program
def main():

# The class 'Pet' is made and will be associated with attributes (name, animal_type, age) and methods 
# (get_name, get_animal_type, get_age)
    class Pet:

# Constructor method is utilized to initialize the object 'my_pet' for 'Pet'. It includes four 
# parameters including 'self', used to modify attributes and access methods.
        def __init__(self,name,animal_type,age):

# The value of the 'name' parameter is passed to the 'name' attribute of Pet        
            self.name = name
# The value of the 'animal_type' parameter is passed to the 'animal_type' attribute of Pet       
            self.animal_type = animal_type
# The value of the 'age' parameter is passed to the 'age' attribute of Pet        
            self.age = age

# function 'set_name' has the user input a name for the pet, the value is then assigned to the 'name'
# attribute of 'Pet'
        def set_name(self):
            self.name = input("What is the name of your pet?: ")

# function 'set_animal_type' has the user input the animal's type for the pet, the value is then assigned to the 'animal_type'
# attribute of 'Pet'
        def set_animal_type(self):
            self.animal_type = input("What is the type of pet? (Bird, Cat, Etc): ")

# function 'set_age' has the user input an age for the pet, the value is then assigned to the 'age'
# attribute of 'Pet'        
        def set_age(self):
            self.age = input("How old is your pet?: ")

# function 'get_name' retrieves and returns a string and the value that was given to the 'name' attribute of 'Pet' 
        def get_name(self):
            return "Pet's Name: " + self.name
        
# function 'get_animal_type' retrieves and returns a string and the value assigned to the 'animal_type' attribute of 'Pet' 
        def get_animal_type(self):
            return "Pet's Type: " + self.animal_type

# function 'get_age' retrieves and returns a string and the value that was given to the 'age' attribute of 'Pet'
        def get_age(self):
            return "Pet's Age: " + self.age

# Program creates object 'my_pet' and the 'Pet' class has three empty strings strings for the 
# initial values of 'name','animal_type', and 'age' parameters
    my_pet = Pet("","","")

# Function 'set_name' is called and assigns the input value to 'name', which is then stored in 'my_pet'
    my_pet.set_name()
# Function 'set_animal_type' is called and assigns the input value to 'animal_type', which is then stored in 'my_pet'
    my_pet.set_animal_type()
# Function 'set_age' is called and assigns the input value to 'age', which is then stored in 'my_pet'
    my_pet.set_age()

# An empty print statement is used to create an empty line for clarity
    print()

# Print statement displays the string 'Pet Information'
    print("[Pet Information]\n")
# Print statement calls the '.get_name()' function for 'my_pet' which displays the 'name' attribute
    print(my_pet.get_name())
# Print statement calls the '.get_animal_type()' function for 'my_pet' which displays the 'animal_type' attribute
    print(my_pet.get_animal_type())
# Print statement calls the '.get_age()' function for 'my_pet' which displays the 'age' attribute
    print(my_pet.get_age())

# The main function will only be called if this script is ran directly    
if __name__ == "__main__":
    main()
```
